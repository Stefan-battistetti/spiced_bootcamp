```python
import pandas as pd
df=pd.read_excel('gapminder_lifeexpectancy.xlsx', index_col=0)
df.head()
```

```python
lifeexp = df[[1950, 1975, 2000, 2015]]
lifeexp.head()
```

```python
import matplotlib.pyplot as plt

lifeexp[2015].hist()
#plt.savefig('histo.png')
```

```python
lifeexp[2015].hist(bins=5)
lifeexp[2015].hist(bins=10)
lifeexp[2015].hist(bins=20)
plt.title('life expectancy in year 2015')
plt.xlabel('age')
plt.ylabel('frequency')
plt.axis([40.0, 90.0, 1.0, 80.0])
```

```python
lifeexp[2015].hist(bins=20, facecolor='#ff0000', alpha=0.5, histtype='bar')
lifeexp[2015].hist(bins=10, facecolor='b', alpha=0.6, histtype='bar')
plt.title('life expectancy in year 2015')
plt.xlabel('age')
plt.ylabel('frequency')
plt.savefig('histo.png', dpi=300)
```
