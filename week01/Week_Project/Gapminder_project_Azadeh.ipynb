Make dataframe long format and check the shape

---



```python
fert = fert.melt(id_vars='country', var_name='year', value_name='fertility_rate')
life =life.melt(id_vars='country', var_name='year', value_name='life_expectancy')
population =population.melt(id_vars='country', var_name='year', value_name='population')
```

merge the 3 dataframes

---



```python
df = fert.merge(life, on=['country','year']).merge(population, on=['country','year'])
```

```python
df
```

```python
df.info()
```

omit NaNs

---



```python
df.dropna(inplace=True)
print(df.shape)
df.head()
```

Add the continets to use as hue for graph


---



```python
continents=pd.read_csv('continents.csv', sep=';',index_col=0)
continents.info()
```

```python
continents.head()
```

```python
continents.reset_index(inplace=True)
continents.head()
```

merge the continent to the existing datafarme

---



```python
df2 = df.merge(continents, how='inner', on='country')
df2.head()
```

```python
df2.tail()
```

Plot 

---



prepare data for plot years between 1960 and 2015

```python
data=df2[(df2['year']>=1960) & (df2['year']<=2015)]
data.head()
```

```python
ymin=data['fertility_rate'].min()
ymax=data['fertility_rate'].max()
xmin=data['life_expectancy'].min()
xmax=data['life_expectancy'].max()
```

**Test: plot only one year with annotations for specific countries**

---



```python
  i=1960
  
  data_1=data[data['year']==i]
  plt.figure(figsize= (12,8))
  sns.set_theme(style="white")
  sns.scatterplot(data = data_1, x = 'life_expectancy', y= 'fertility_rate', size='population', sizes=(200,1000), hue='continent', alpha=0.7, palette='muted')
  
  plt.legend(bbox_to_anchor=(1.02, 1), loc=2)
  plt.ylabel('Fertility Rate', fontsize = 14)
  plt.xlabel('Life Expectancy [in years]', fontsize = 14)
  plt.title('Life quality in ' + str(i), fontsize = 20)
  plt.axis((xmin-2, xmax+2, ymin-2, ymax+2))

  # Annotate data
  # prepration for anotating data in graph
  China = data[(data['year']==i) & (data['country'] == 'China')]
  India =data[(data['year']==i) & (data['country'] == 'India')]
  USA=data[(data['year']==i) & (data['country'] == 'United States')]
  Germany =data[(data['year']==i) & (data['country'] == 'Germany')]

  plt.text(India['life_expectancy']+1.1,India['fertility_rate'], 'India')
  plt.text(China['life_expectancy']+1.1,China['fertility_rate'], 'China')
  plt.text(USA['life_expectancy']+1.1,USA['fertility_rate'], 'USA')
  plt.text(Germany['life_expectancy']+1.1,Germany['fertility_rate'], 'Germany') 
```

**  make graphs for all years and Annotate data for specific countries**
---



```python
# make graphs for all years and Annotate data for specific countries
for i in range (1960,1962):
  data_1=data[data['year']==i]
  plt.figure(figsize= (12,8))
  sns.set_theme(style="white")
  sns.scatterplot(data = data_1, x = 'life_expectancy', y= 'fertility_rate', size='population', sizes=(20,1000), hue='continent', alpha=0.7, palette='muted')
  
  plt.legend(bbox_to_anchor=(1.02, 1), loc=2)
  plt.ylabel('Fertility Rate', fontsize = 14)
  plt.xlabel('Life Expectancy [in years]', fontsize = 14)
  plt.title('Life quality in ' + str(i), fontsize = 20)
  plt.axis((xmin-2, xmax+2, ymin-2, ymax+2))

  # Annotate data for specific countries
  # prepration for anotating data in graph
  China = data[(data['year']==i) & (data['country'] == 'China')]
  India =data[(data['year']==i) & (data['country'] == 'India')]
  USA=data[(data['year']==i) & (data['country'] == 'United States')]
  Germany =data[(data['year']==i) & (data['country'] == 'Germany')]

  plt.text(India['life_expectancy']+1.1,India['fertility_rate'], 'India')
  plt.text(China['life_expectancy']+1.1,China['fertility_rate'], 'China')
  plt.text(USA['life_expectancy']+1.1,USA['fertility_rate'], 'USA')
  plt.text(Germany['life_expectancy']+1.1,Germany['fertility_rate'], 'Germany')
  
  plt.savefig('graph'+str(i)+'.png', dpi=100, bbox_inches='tight')
  plt.close() 
```

**Annotate data all the countries**


---



```python
# make graphs for all years for all years
for i in range (1960,2015):
  data_1=data[data['year']==i]
  plt.figure(figsize= (25,15))
  sns.set_theme(style="white")
  sns.scatterplot(data = data_1, x = 'life_expectancy', y= 'fertility_rate', size='population', sizes=(20,1000), hue='continent', alpha=0.7, palette='muted')
  
  plt.legend(bbox_to_anchor=(1.02, 1), loc=2)
  plt.ylabel('Fertility Rate', fontsize = 14)
  plt.xlabel('Life Expectancy [in years]', fontsize = 14)
  plt.title('Life quality in ' + str(i), fontsize = 20)
  plt.axis((xmin-2, xmax+2, ymin-2, ymax+2))

  # Annotate data all the countries 
  for j in range (data_1.shape[0]):
    plt.text(data_1.iloc[j,3]+1.1,data_1.iloc[j,2], data_1.iloc[j,0])
  
  plt.savefig('graph2_'+str(i)+'.png', dpi=100, bbox_inches='tight')
  plt.close() 
```
