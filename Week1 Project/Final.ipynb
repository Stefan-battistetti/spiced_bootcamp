```python
# Project Week 1

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fert = pd.read_csv('gapminder_total_fertility.csv', index_col=0)
life = pd.read_excel('gapminder_lifeexpectancy.xlsx', index_col=0)
pop  = pd.read_excel('gapminder_population.xlsx', index_col=0)
con  = pd.read_csv('continents.csv', sep=';')
```

```python
fert = fert.reset_index()
fert.rename(columns={'Total fertility rate': 'country'}, inplace=True)
fert.head()
```

```python
pop = pop.reset_index()
pop.rename(columns={'Total population': 'country'}, inplace=True)
pop.head()
```

```python
life = life.reset_index()
life.rename(columns={'Life expectancy': 'country'}, inplace=True)
life.head()
```

```python
tidy_fert = pd.melt(fert,
                         id_vars='country',
                         var_name='year',
                         value_name='Total fertility rate')
tidy_fert.head()
```

```python
tidy_life = pd.melt(life,
                         id_vars='country',
                         var_name='year',
                         value_name='life expectancy')
tidy_life.head()
pop.head()
```

```python
tidy_pop = pd.melt(pop,
                         id_vars='country',
                         var_name='year',
                         value_name='Total population')
tidy_pop.head()
```

```python
df = tidy_fert.merge(tidy_pop)
df.head()
```

```python
df_2 = df.merge(tidy_life)
df_2.head()
con.head()
```

```python
df_final = df_2.merge(con, on='country', how='inner')
df_final.head()
df_final.rename
```

```python
for i in range(1960,2016):
    df_fi = df_final.loc[df_final['year'] == int(i)]
    sns.relplot(x='life expectancy', y='Total fertility rate', hue='continent', size='Total population', data=df_fi, alpha=0.6, sizes=(10,500))
    plt.title('Year' + str(i), y=0.95, loc='right')
    plt.savefig(str(i) + '.png')
    plt.close()
   
```
