```python
import pandas as pd
import imageio
fert=pd.read_csv("gapminder_total_fertility.csv", index_col=0)
fert.head()
```

```python
life=pd.read_excel("gapminder_lifeexpectancy.xlsx", index_col=0)
```

```python
print(life.shape)
print(fert.shape)
```

```python
fert.columns
```

```python
life.columns
```

```python
fert.columns=fert.columns.astype(int)
fert.columns
```

```python
fert.index
fert.index.name='country'
fert.head()
```

```python
fert=fert.reset_index()
```

```python
fert=fert.melt(id_vars='country', var_name='year', value_name='fertility_rate')
fert.head()
```

```python
life.index
life.head()
```

```python
life.index.name='country'
life=life.reset_index()
life.head()
```

```python
life=life.melt(id_vars='country',var_name='year', value_name='life_expectancy')
```

```python
life.head()
```

```python
pop=pd.read_excel('gapminder_population.xlsx', index_col=0)
```

```python
pop.index
pop.index.name='country'
pop=pop.reset_index()
pop.head()
```

```python
pop=pop.melt(id_vars='country', var_name='year', value_name='population')
pop.head()
```

```python
df=fert.merge(pop).merge(life)
df.dropna()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_subset=df.loc[df['year']==2000]
sns.scatterplot(x='life_expectancy', y='fertility_rate', data=df_subset, alpha=0.6, hue='population')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_subset=df.loc[df['year']==2000]
sns.scatterplot(x='fertility_rate', y='life_expectancy', data=df_subset, alpha=0.6, hue='population', size='population', sizes=(50,500))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_subset=df.loc[df['year']==2000]
sns.scatterplot(x='fertility_rate', y='life_expectancy', data=df_subset, alpha=1, hue='population', size='population', sizes=(100,8000))
sns.set(rc={'figure.figsize':(10,15)})
```

```python
for i in range(1960,2016):
    df_subset=df.loc[df['year']==i]
    sns.scatterplot(x='fertility_rate', y='life_expectancy', data=df_subset, alpha=1, 
                    hue='population', size='population', sizes=(100,8000)).set(title='Year'+'  '+str(i))
    sns.set(rc={'figure.figsize':(10,10)})
    plt.savefig('lifeexp_' + str(i) + '.png')
    plt.close()
    
```
