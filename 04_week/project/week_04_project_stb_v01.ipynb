**Next step: Scrape the lyrics from the website**

```python
lyrics = []
def make_lyrics_list(urls):
  for i in urls:
    response = requests.get(i)
    artist_html = response.text
    artist_soup = BeautifulSoup(artist_html, 'html.parser')
    
    ds_curriculum = artist_soup.find_all(class_ ='lyric-body')
    #print(ds_curriculum)
    #lyrics.append(ds_curriculum)
    for tags in ds_curriculum:
      #print(tags.name)
      lyrics.append(tags.text)
    


```

At the end I have to combine the two dataframes which I created out of each corpus

The dataframe needs to have an additional row called "artists", which needs to be endcoded as 1 and 2, since the model needs to predict the artits, which is at the end the "y"



```python

def make_dataframe(url):
    '''
   BeautifulSoup 
    '''
    response = requests.get(url)
    spiced_html = response.text
    spiced_soup = BeautifulSoup(spiced_html)
    ds_curriculum = spiced_soup.find_all(class_ = 'curriculum-mini-section')
    topic = []
    description = []

    for items in ds_curriculum:
        topic.append(items.h3.text)
        description.append(items.find(class_ = "mob-hidden").text)
    
    dataframe = pd.DataFrame({'topic':topic, 'description':description})

    return dataframe
```
