```python
###create scatter plot
sns.scatterplot(x='life expectancy', y='fertility_rate', data=df_2015, alpha=0.6)

plt.savefig('fil_2015.png')
```

```python
df_all
```

```python
xmin = df_all['life expectancy'].min()
xmax = df_all['life expectancy'].max()
ymin = df_all['fertility_rate'].min()
ymax = df_all['fertility_rate'].max()
```

```python
df_all.info()
```

```python
df_all['year'] = df_all['year'].astype(int)
```

```python
df_all.info()
```

```python
for i in range(1960,2015):
    df_sub = df_all.loc[df_all['year']==i]
    ###print(df_sub.shape)
    sns.scatterplot(x='life expectancy', y='fertility_rate',
                    size='population', hue= 'country', sizes = (30, 400), legend=False, 
                    data=df_sub, alpha=0.6)
    plt.axis = ((xmin, xmax, ymin, ymax))
    plt.savefig('lifeexp_' + str(i)+'.png')
    plt.close()
    

```

```python
import imageio
images = []

for r in range(1960,2015):
    filename = 'lifeexp_{}.png'.format(r)
    images.append(imageio.imread(filename))
imageio.mimsave('output.gif', images, fps=20)
```
