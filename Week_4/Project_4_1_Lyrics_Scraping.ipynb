```python
import requests
import re
import pprint
from bs4 import BeautifulSoup
import pickle  ###safe list using pickle
```

Use the requests module to download that page


```python
URL = requests.get('https://www.lyrics.com/artist/Selena-Gomez/1063097').text
###including .text in the code is alraedy attributing data as a text file
```

Save the page to a text file

```python
with open ("URL", "w", encoding='utf-8') as f:   ##use 'w' to write to file
    f.write(URL)
```

Extract all links using Regular Expressions

```python
#re.findall('<.+?>', text)
#lyric/35263516/Selena+Gome
pattern = "lyric/\d+/Selena\+Gomez/[^\"]+"
Selena_lyrics=re.findall(pattern, URL, re.IGNORECASE)  ###findall results in s
```

```python
pprint.pprint(Selena_lyrics)
```

Write a loop that goes through all song to include the URL link

```python
for song in range(len(Selena_lyrics)):
    Selena_lyrics[song] = 'https://www.lyrics.com/' + Selena_lyrics[song]   
   
```

```python
pprint.pprint(Selena_lyrics)
```

```python
print(len(Selena_lyrics))
```

Run a loop to create coitus of the lyrics

```python
p=[]

for lyrics in Selena_lyrics:
    check = requests.head(lyrics)
    status_code= check.status_code
    if status_code==200:                 ###if statement to exclude unsuccesfull url links from the loop
        All_sel = requests.get(lyrics)   ##get the lyrics
        All_sel_txt = All_sel.text       ##convert lyrics to txt file
        sel_soup = BeautifulSoup(All_sel_txt, 'html.parser')    ###convert url to beautiful soup
        sel_all= sel_soup.find(attrs={"class":"lyric-body"}).text   ###scarpe lyriic using beautiful soup
        p.append(sel_all)    ###append the scrape file to the empty 'p' list
    else:
        continue
  
    #Emin_all= Emin_soup.find(id="lyric-body-text").text
        #t = lyrics.split('/')[-1]
    #print(t)
        #with open (t + '.txt', "w") as f:
            #f.write(Emin_all)
```

```python
###save the list
pickle_out =open('selena_lyrics_all', 'wb')   ###create file name to safe list, 'wb' means write as byte
pickle.dump(p, pickle_out)                    ##state the object that need to be save in file
pickle_out.close()                            ###close the operation
```

# Download Tiwa Savage

```python
##get Tiwa URL
URL_T = requests.get('https://www.lyrics.com/artist/Tiwa-Savage/2237809').text
 #/2342870  
```

```python
pprint.pprint(URL_T)
```

```python
pattern = "lyric/\d+/Tiwa\+Savage/[^\"]+"
Tiwa_lyrics=re.findall(pattern, URL_T, re.IGNORECASE) 
```

```python
pprint.pprint(Tiwa_lyrics)
```

```python
for song in range(len(Tiwa_lyrics)):
    Tiwa_lyrics[song] = 'https://www.lyrics.com/' + Tiwa_lyrics[song]  
```

```python
pprint.pprint(Tiwa_lyrics)
```

```python
T=[]

for lyrics in Tiwa_lyrics:
    check = requests.head(lyrics)
    status_code= check.status_code
    if status_code==200:   ###check if the there is an error code then skip
        All_Tiwa = requests.get(lyrics)
        All_Tiwa_txt = All_Tiwa.text
        Tiwa_soup = BeautifulSoup(All_Tiwa_txt, 'html.parser')
        Tiwa_all= Tiwa_soup.find(attrs={"class":"lyric-body"}).text
        T.append(Tiwa_all)
    else:
        continue
```

```python
###safe tiwa song
pickle_out =open('Tiwa_lyrics_all', 'wb')  ##name of file, and wb save as bytes
pickle.dump(T, pickle_out)   ###file that need to be written
pickle_out.close()   ##close the operation
```

```python
##get Tiwa URL
URL_ed = requests.get('https://www.lyrics.com/artist/Ed-Sheeran').text
```
