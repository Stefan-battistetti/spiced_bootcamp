```python
import pickle
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
import pprint
```

```python
###Load selena lyrics
pickle_in =open('selena_lyrics_all', 'rb')
selena_copus=pickle.load(pickle_in)
pickle_in =open('Ed_lyrics_all', 'rb')
Ed_copus=pickle.load(pickle_in)
```

```python
print(len(selena_copus))
print(len(Ed_copus))
```

```python
##merge lyrics
prep_copus=selena_copus + Ed_copus ##corpus is X
labels= ['selena']*569 + ['Ed_sheren']*601  ##labels is y
```

Preprocessing of the copus(lyrics)

```python
##replace r\n, r in the list and replace with white space
prep_copus = [elements.replace('\r\n', ' ').replace('\n',' ') for elements in prep_copus]
```

```python
import neattext as nfx

prep_copus_df=pd.DataFrame(prep_copus,index=labels)  ##neattext only work on data frame
prep_copus_df.head(5)
```

```python
prep_copus_df[0]=prep_copus_df[0].apply(nfx.remove_stopwords)   ###remove stop words
prep_copus_df[0]=prep_copus_df[0].apply(nfx.remove_numbers)     ##remove numbers
prep_copus_df[0]=prep_copus_df[0].apply(nfx.remove_special_characters) ##rem special characters
prep_copus_df[0]=prep_copus_df[0].apply(nfx.remove_punctuations)   ###remo punctuations
prep_copus_df[0]=prep_copus_df[0].str.lower()
```

```python
prep_copus_df.head(5)
```

```python
##nltk.download("wordnet")  ###needs to be run once
```

```python
import nltk
from nltk.tokenize import TreebankWordTokenizer 
from nltk.stem import WordNetLemmatizer
```

```python
#Preprocess unsing nltk
#CORPUS = [s.lower() for s in prep_corpus_df
CLEAN_CORPUS = []

tokenizer = TreebankWordTokenizer()
lemmatizer = WordNetLemmatizer()


for doc in prep_copus_df[0]:
    tokens = tokenizer.tokenize(text=doc)
    clean_doc = " ".join(lemmatizer.lemmatize(token) for token in tokens)
    CLEAN_CORPUS.append(clean_doc)


```

```python
pprint.pprint(CLEAN_CORPUS)
```

```python
###split data to test and Train
##stratify use to get same proportion of dataset
X_train, X_test, y_train, y_test = train_test_split(CLEAN_CORPUS, labels, random_state=10)  
```

```python
print(len(X_train))
print(len(y_train))
print(len(X_test))
print(len(y_test))
```

```python
from nltk.corpus import stopwords
STOPWORDS = stopwords.words('english')

vectorizer=TfidfVectorizer(stop_words=STOPWORDS)

X_train_vex=vectorizer.fit_transform(X_train)  ##train

X_test_vex=vectorizer.transform(X_test)  ##testn
```

```python
###creat data frame for the Xtrain and Xtest
##just to view dataset
X_train_vex_df=pd.DataFrame(X_train_vex.todense(), columns=vectorizer.get_feature_names_out(), index=y_train)

X_train_vex_df.head(5)  ###to view dataset
```

Fit the classification model using Multinomial NB model

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB()
model.fit(X_train_vex, y_train)
```

```python
model.score(X_train_vex, y_train)
```

Predict on test data

```python
model.score(X_test_vex, y_test)
```

```python
model.predict_proba(X_test_vex)
```

```python
y_pred = model.predict(X_test_vex)
```

Evaluate model

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
print(accuracy_score(y_test, y_pred))
print(recall_score(y_test, y_pred, pos_label='Ed_sheren'))
print(precision_score(y_test, y_pred, pos_label='Ed_sheren'))
print(f1_score(y_test, y_pred, pos_label='Ed_sheren'))
```

```python
print(accuracy_score(y_test, y_pred))
print(recall_score(y_test, y_pred, pos_label='selena'))
print(precision_score(y_test, y_pred, pos_label='selena'))
print(f1_score(y_test, y_pred, pos_label='selena'))
```
