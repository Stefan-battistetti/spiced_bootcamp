```python
plt.figure(dpi=150)

for i in range (1960, 2015):
  df_subset = df.loc[df['year']==i]
  np_pop = np.array(df_subset['population'])
  np_pop2 = np_pop*2
  plt.figure(dpi=150)
  sns.scatterplot(x='life_expectancy', y='fertility_rate', hue =df_subset['population'],data=df_subset, alpha=0.6, size=np_pop2, sizes=(20,400), legend = False)
  #sns.scatterplot(df_subset['population'], hue = df_subset['population'], size=np_pop2, sizes=(20,400), legend = False)
  plt.axis([x_min, x_max, y_min, y_max])
  plt.grid(True)
  plt.savefig('lifeexp_'+ str(i))
  #files.download('lifeexp_'+ str(i))
  plt.show()
  plt.close()
plt.savefig('lifeexp_'+ str(i))
  
  
```
