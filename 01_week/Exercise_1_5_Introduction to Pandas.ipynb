## Draw a histogram

```python
lifeexp = pd.read_excel('gapminder_lifeexpectancy.xlsx', index_col = 0)
```

```python
lifeexp.head()
```

```python
lifeexp = lifeexp[[1950, 1975, 2000, 2015]]
```

```python
import matplotlib.pyplot as plt

lifeexp[2015].hist()
plt.savefig('histo.png')
```

```python
lifeexp[2000].hist(bins=5)
lifeexp[2000].hist(bins=10)
lifeexp[2000].hist(bins=20)
plt.axis([30.0, 90.0, 0.0, 90.0])
facecolor='#ff0000',
alpha=0.75,
histtype='bar',
plt.savefig('histo.png', dpi=150)
plt.savefig('histo.svg', dpi=150)
```
