```python
df1 = fert.merge(population, how='inner', on=['country','year'])
df1
```

```python
df = df1.merge(life, how='inner', on=['country','year'])
df
```

```python
df_subset = df.loc[df['year'] == 2000]
sns.scatterplot(x='life_expectancy', y='fertility_rate',
            data=df_subset, alpha=0.6) #size='population', sizes=(10,200)
```

```python
for i in range(1960, 2016):
    p = sns.scatterplot(x='life_expectancy', y='fertility_rate',
    data=df.loc[df['year']== i], alpha=0.6, size='population', sizes=(10,200), hue='country', legend= False)
    plt.title(i)
    plt.axis([0, 100, 0, 8]) #to fix the coordinates of the plot.
    plt.savefig('lifeexp_' + str(i) + '.png')
    plt.close() 
    
```

```python
images = []

for i in range(1960, 2016):
    filename = 'lifeexp_{}.png'.format(i)
    images.append(imageio.imread(filename))

imageio.mimsave('output.gif', images, fps=20)
```

```python
cont = pd.read_csv('continents.csv', sep = ';')
cont
```

```python
df_c = df.merge(cont, how='inner', on='country')
df_c
```

```python
conta = df_c.groupby(['continent', 'year'])[['life_expectancy', 'fertility_rate']].mean()
conta
```

```python
conta2 = conta.rename(columns={'life_expectancy': 'avg.life_expectancy', 'fertility_rate': 'avg.fertility_rate'})
```

```python
df_f1 = df_c.merge(conta2, how='inner', on=['continent', 'year'])
df_f1
```

```python
for i in range(1960, 2016):
    p = sns.scatterplot(x='life_expectancy', y='fertility_rate',
    data=df_f1.loc[df_f1['year']== i], alpha=0.6, size='population', sizes=(10,200), hue='continent', legend= False)
    pa = sns.scatterplot(x='avg.life_expectancy', y='avg.fertility_rate',
    data=df_f1.loc[df_f1['year']== i], alpha=0.6, marker = '+', hue='continent', legend= 'full')


    #plt.legend(labels=('continent'))
    plt.title(i)
    plt.axis([0, 100, 0, 8]) #to fix the coordinates of the plot.
    plt.savefig('lifeexp_' + str(i) + '.png')
    plt.close() 
```
