```python
#load the files

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

fert = pd.read_csv('gapminder_total_fertility.csv')
life = pd.read_excel('gapminder_lifeexpectancy.xlsx')
pop = pd.read_excel('gapminder_population.xlsx')
pop = pop.dropna(axis = 1, how = 'all')
```

```python
#change the name of the countries column

fert.rename(columns={'Total fertility rate': 'country'}, inplace=True)
life.rename({'Life expectancy': 'country'}, axis ='columns', inplace=True)
pop.rename(columns={'Total population': 'country'}, inplace=True)
```

```python
#shape the dataframes to a long format

tidy_fert = pd.melt(fert,                        
                    id_vars='country',           
                    var_name='year'  ,           
                    value_name='Fertility_rate')

tidy_life = pd.melt(life,                        
                    id_vars='country',           
                    var_name='year'  ,           
                    value_name='Life_expectancy')

tidy_pop = pd.melt(pop,                        
                    id_vars='country',           
                    var_name='year'  ,           
                    value_name='Total_population')
```

```python
tidy_fert["year"] = tidy_fert["year"].astype(int)
```

```python
tidy_fert.info()
```

```python
#merge data frames

final = tidy_fert.merge(tidy_life, left_on=None, right_on=None, left_index=False, right_index=False)
final = final.merge(tidy_pop, left_on=None, right_on=None, left_index=False, right_index=False)
```

```python
final.head(3)
```

```python
final.describe()
```

```python
#selection of column and plot

final = final.loc[final['year'] == 1990]
sns.scatterplot(x='Life_expectancy', 
                y = 'Fertility_rate', 
                data = final, 
                alpha = 0.6, 
                hue = 'country', 
                style='country', 
                size = 'Total_population', 
                legend = False)
plt.axis((40, 80, 1, 8))
plt.title('Life expectancy vs. Fertility rate')
```

```python
#scatterplots showing the correlation between life expectancy and fertility (1960 - 2015)

for i in range(1960, 2016):
    final_scatter = final.loc[final['year'] == i
    sns.scatterplot(x='Life_expectancy', 
                y = 'Fertility_rate', 
                data = final, 
                alpha = 0.6, 
                hue = 'country', 
                style='country', 
                size = 'Total_population', 
                legend = False)
    filename = 'lifeexp_{}.png'.format(i)
    plt.savefig(filename)
    plt.close()
```

```python
#generating a gif file from the individual pictures

import imageio

images = []

for i in range(1960, 2016):
    filename = 'lifeexp_{}.png'.format(i)
    images.append(imageio.imread(filename))

imageio.mimsave('output.gif', images, fps=20)
```
