```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tf_vec = TfidfVectorizer()
```

```python
X = psb_kylie.lyrics
y = psb_kylie['Artist_Name']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 42, test_size = 0.2)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
tf_vec.fit(X_train)
```

```python
X_train_trans = tf_vec.transform(X_train).todense()
```

```python
X_test_trans = tf_vec.transform(X_test).todense()
```

```python
print(tf_vec.vocabulary_)
```

```python
len(tf_vec.vocabulary_)
```

```python
from sklearn.linear_model import LogisticRegression
import numpy as np
```

```python
m = LogisticRegression()
```

```python
m.fit(np.array(X_train_trans), y_train)
```

```python
['text psb text kylie']
```

```python
y1_pred = m.predict(np.asarray(X_test_trans))
y1_pred
```

```python
m.predict(tf_vec.transform(['text psb text kylie']))
```

```python
m.predict_proba(tf_vec.transform(['text psb text kylie']))
```

```python
round(m.score(np.asarray(X_train_trans), y_train),3)
```

```python
round(m.score(np.asarray(X_test_trans), y_test),3)
```
