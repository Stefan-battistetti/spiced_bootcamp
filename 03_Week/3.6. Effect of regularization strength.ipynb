```python
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

X, y = load_breast_cancer(True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state= 42)
X_train.shape , X_test.shape , y_train.shape , y_test.shape
```

```python
m = LogisticRegression()
m.fit(X_train, y_train)
```

```python
ypred = m.predict(X_test)
```

```python
print(m.score(X_train, y_train))

print(m.coef_[0])
```

```python
coef_ = pd.DataFrame(m.coef_.T, columns=['Coefficients'])
coef_.plot(kind='barh', figsize=(9, 7))
plt.title('model coefficients')
plt.axvline(x=0, color='.5', linestyle='--')
plt.show()
```

```python
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
```

```python
m_ridge = Ridge()
m_lasso = Lasso()

m_ridge.fit(X, y)  
m_lasso.fit(X, y)  
 
ypred_ridge = m_ridge.predict(X)
ypred_lasso = m_lasso.predict(X)
```

```python
ridge_coefs = pd.DataFrame(m_ridge.coef_.T, columns=['Coefficients'])
ridge_coefs.plot(kind='barh', figsize=(9, 7))
plt.title('Ridge model coefficients')
plt.axvline(x=0, color='.5', linestyle='--')
plt.show()
```

```python
lasso_coefs = pd.DataFrame(m_lasso.coef_.T, columns=['Coefficients'])
lasso_coefs.plot(kind='barh', figsize=(9, 7))
plt.title('Lasso model coefficients')
plt.axvline(x=0, color='.5', linestyle='--')
plt.show()
```
