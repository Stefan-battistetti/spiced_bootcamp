```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('train.csv', parse_dates=True)
df.info()
```

```python
ts = pd.to_datetime(df['datetime'])
df['year'] = ts.dt.year
df['month'] = ts.dt.month
df['week'] = ts.dt.week
df['day'] = ts.dt.day
df['hour'] = ts.dt.hour
```

```python
sns.lineplot(df['year'],df['count'])
```

```python
sns.lineplot(df['month'],df['count'])
```

```python
sns.lineplot(df['week'],df['count'])
```

```python
sns.lineplot(df['day'],df['count'])
```

```python
sns.lineplot(df['hour'],df['count'])
```
