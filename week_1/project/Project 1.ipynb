```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import imageio
```

```python
fert = pd.read_csv("gapminder_total_fertility.csv")
life = pd.read_excel("gapminder_lifeexpectancy.xlsx")
pop = pd.read_excel("gapminder_population.xlsx")
continents = pd.read_csv("continents.csv", sep=';')
```

```python
def rename_column(df, column_name, new_name):
    """take a column in str of a dataframe and return the column with a new name"""
    df.rename(columns={column_name:new_name}, inplace=True)
    return df.head()
```

```python
def change_type_to_int(df, column):
    """change the type of a dataframe column to int"""
    df[column] = df[column].astype(int)
```

```python
rename_column(pop, 'Total population', 'country')
```

__Change the format of the dataframes__

```python
fert = pd.melt(fert, id_vars='country', var_name='year', value_name='fertility_rate')
```

```python
life = pd.melt(life, id_vars = 'country', var_name='year', value_name='life_expectancy')
```

```python
pop = pd.melt(pop, id_vars = 'country', var_name='year', value_name='population size')
```

__Change type of columns "year" to int__

```python
change_type_to_int(fert, 'year')
```

```python
change_type_to_int(life, 'year')
```

```python
change_type_to_int(pop, 'year')
```

__merge all the dataframes__

```python
df_f_p = fert.merge(pop, how='inner', on=['country', 'year'])
```

```python
df = pd.merge(df_f_p, life, on=['country', 'year'])
```

```python
df = pd.merge(df, continents, on='country')
```

```python
df.head()
```

__Check for missing values and drop them__

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

__Create a subset df for 2000 and plot it__

```python
df_subset = df.loc[df['year'] == 2000]
```

```python
#sns.scatterplot(x='life_expectancy', y='fertility_rate',
            #data=df_subset, alpha=0.6, )
fig, axis = plt.subplots(figsize=(15,10))
sns.scatterplot(x='life_expectancy', y='fertility_rate',
                    hue = 'continent',size='population size', legend = False, palette="deep",
                    data=df_subset, alpha=0.6, ax=axis, sizes=(100, 500))

plt.show()
```

__Find the min and max values of life expectancy and fertility rate__

```python
xmin = df['life_expectancy'].min()
xmax = df['life_expectancy'].max()
ymin = df['fertility_rate'].min() 
ymax= df['fertility_rate'].max()

list_of_min_and_max = [xmin, xmax, ymin, ymax]
print(list_of_min_and_max)
```

__Create a for loop for generating scatterplots between 1960 and 2015__

```python
for year in range(1960,2016):
    df_subset = df.loc[df['year'] == year]
    
    fig, axis = plt.subplots(figsize=(15,10))
    g = sns.scatterplot(x='life_expectancy', y='fertility_rate',
                    hue = 'continent',size='population size', palette="deep",
                    data=df_subset, alpha=0.6, ax=axis, sizes=(100, 500), hue_norm=(0, 7))
    plt.title(year)
    h,l = g.get_legend_handles_labels()
    plt.legend(h[0:8],l[0:7],loc = 'upper right', borderaxespad=0.)
    plt.axis([3, 85, 0.5, 10])
    plt.savefig('lifeexp_' + str(year) + '.png')
    plt.close()


```

__Create the gif__

```python
images = []

for i in range(1960,2016):
    filename = 'lifeexp_{}.png'.format(i)
    images.append(imageio.imread(filename))

imageio.mimsave('output.gif', images, fps=20)
```
